```python
tips.head()
```

```python
sns.set_style('whitegrid')

plt.figure(dpi=100, figsize=(10, 4))
ax = sns.boxplot(data=tips, x='PRS',y='score',width=0.6, hue='label')
ax.set_ylabel('Score', fontsize=10)
ax.xaxis.label.set_visible(False)
ax.yaxis.label.set_visible(True)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5'], fontsize=10)

#plt.show()
plt.savefig("D:/desk/JNU/fig/clusterscore.jpg", dpi = 600)
```
